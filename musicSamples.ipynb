```python
import pickle
import os
import random
from music21 import *
```

```python
path = 'music_samples.pkl'

# store music samples into musicSamples array
with open(path, 'rb') as file:
    musicSamples = pickle.load(file)
```

```python
# make folder to store created sample MIDI files
from music21 import note

folder_name = 'sampleMidiFiles'
os.makedirs(folder_name, exist_ok=True) 

for i in range(len(musicSamples)):
    noteStream = stream.Stream()
    notesSample = musicSamples[i]

    for n in notesSample:
        noteObj = note.Note(n)
        noteStream.append(noteObj)

    noteStream.append(tempo.MetronomeMark(number=120))
    midi_file_path = os.path.join(folder_name, f'test_{i}.mid')
    midiFile = noteStream.write('midi', fp=midi_file_path)
```

```python
folder_name = 'sampleMidiFiles'
midiFiles = [file for file in os.listdir(folder_name) if file.endswith('.mid')]

#randomly select 10 MIDI files
selectedFiles = random.sample(midiFiles, 10)

for file in selectedFiles:
    file_path = os.path.join(folder_name, file)
    
    # print file name
    print(f"Displaying musical notation for: {file}")
    
    midiStream = converter.parse(file_path)
    
    # show musical notation + audio sample
    midiStream.show()
    midiStream.show("midi")
```
